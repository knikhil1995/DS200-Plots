```python
import matplotlib.pyplot as plt
import numpy as np


numschools=np.genfromtxt('schoolsvsenrollments.csv', delimiter=',', dtype=None,encoding=None)

fig, ax = plt.subplots()

ax.scatter(numschools[1:,1].astype(int),numschools[1:,4].astype(int), label='Primary Schools')
ax.scatter(numschools[1:,2].astype(int),numschools[1:,5].astype(int), label='Middle Schools')
ax.scatter(numschools[1:,3].astype(int),numschools[1:,6].astype(int), label='High/Senior Secondary Schools')

ax.legend()
ax.grid(True)
plt.xlabel("Number of Schools")
plt.ylabel("Student Enrollment")
plt.title("District-wise student enrollment in Himachal Pradesh, 2020-21")
plt.savefig('Scatter.pdf', format='pdf')
```

```python
import matplotlib.pyplot as plt
import numpy as np


rainfall=np.genfromtxt('All-India-Rainfall-Act_Dep_1901_to_2019_0.csv', delimiter=',', dtype=None,encoding=None)

fig, ax = plt.subplots()

ax.boxplot(rainfall[1:,1:5].astype(np.double),labels=["June","July","August","September"])
plt.ylabel("Rainfall in MM")
plt.title("Monthly all India rainfall from 1901 to 2019")
plt.savefig('box.pdf', format='pdf')
```

```python
import matplotlib.pyplot as plt
import numpy as np

ind = np.arange(12) 
enrollment=np.genfromtxt('Enrolmen-in-District-wise-during-2020-21.csv', delimiter=',', dtype=None,encoding=None)

fig, ax = plt.subplots(figsize=(12,5))

p1 = ax.bar(ind, enrollment[1:-1,4].astype(int), tick_label=enrollment[1:-1,1].astype(str), label="Primary School")
p2 = ax.bar(ind, enrollment[1:-1,7].astype(int), tick_label=enrollment[1:-1,1].astype(str),bottom=enrollment[1:-1,4].astype(int),label="Middle School")
p3 = ax.bar(ind, enrollment[1:-1,10].astype(int), tick_label=enrollment[1:-1,1].astype(str),bottom=enrollment[1:-1,4].astype(int)+enrollment[1:-1,7].astype(int),label="High/Senior Secondary School")
ax.legend()
plt.ylabel("Student Enrollment")
plt.title("District-wise student enrollment in Himachal Pradesh, 2020-21")

plt.savefig('bar.pdf', format='pdf')
```
